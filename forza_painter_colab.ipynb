```python
#@title Geometrize Engine
import numpy as np
from concurrent.futures import ProcessPoolExecutor
import multiprocessing
import time
import math

# Shape type constants
RECTANGLE = 1
ROTATED_ELLIPSE = 16

# ---------- GPU Setup ----------
USE_GPU = False
if GPU_AVAILABLE:
    try:
        from numba import cuda
        import numba

        @cuda.jit
        def _gpu_compute_error_kernel(canvas, target, masks_x, masks_y, masks_w, masks_h,
                                       masks_rot, masks_type, colors_r, colors_g, colors_b,
                                       errors_out, img_h, img_w, n_candidates):
            """Compute error reduction for each candidate shape on GPU."""
            idx = cuda.grid(1)
            if idx >= n_candidates:
                return

            cx = masks_x[idx]
            cy = masks_y[idx]
            hw = masks_w[idx]
            hh = masks_h[idx]
            rot = masks_rot[idx]
            stype = masks_type[idx]
            cr = colors_r[idx]
            cg = colors_g[idx]
            cb = colors_b[idx]

            cos_a = math.cos(rot * math.pi / 180.0)
            sin_a = math.sin(rot * math.pi / 180.0)

            error_reduction = 0.0
            x_min = max(0, cx - hw)
            x_max = min(img_w, cx + hw)
            y_min = max(0, cy - hh)
            y_max = min(img_h, cy + hh)

            for py in range(y_min, y_max):
                for px in range(x_min, x_max):
                    # Check if pixel is inside shape
                    dx = px - cx
                    dy = py - cy
                    inside = False
                    if stype == 1:  # Rectangle
                        rx = dx * cos_a + dy * sin_a
                        ry = -dx * sin_a + dy * cos_a
                        if abs(rx) <= hw and abs(ry) <= hh:
                            inside = True
                    else:  # Ellipse
                        rx = dx * cos_a + dy * sin_a
                        ry = -dx * sin_a + dy * cos_a
                        if hw > 0 and hh > 0:
                            val = (rx * rx) / (hw * hw) + (ry * ry) / (hh * hh)
                            if val <= 1.0:
                                inside = True

                    if inside:
                        # Current error
                        old_dr = float(canvas[py, px, 0]) - float(target[py, px, 0])
                        old_dg = float(canvas[py, px, 1]) - float(target[py, px, 1])
                        old_db = float(canvas[py, px, 2]) - float(target[py, px, 2])
                        old_err = old_dr * old_dr + old_dg * old_dg + old_db * old_db
                        # New error with shape color
                        new_dr = float(cr) - float(target[py, px, 0])
                        new_dg = float(cg) - float(target[py, px, 1])
                        new_db = float(cb) - float(target[py, px, 2])
                        new_err = new_dr * new_dr + new_dg * new_dg + new_db * new_db
                        error_reduction += old_err - new_err

            errors_out[idx] = error_reduction

        USE_GPU = True
        print("GPU acceleration enabled with numba.cuda")
    except Exception as e:
        print(f"GPU setup failed ({e}), falling back to CPU.")
        USE_GPU = False

if not USE_GPU:
    print("Using CPU-only mode with numpy vectorized operations and multiprocessing.")


def posterize_color(r, g, b, levels):
    """Quantize color to the given number of posterization levels."""
    if levels <= 1:
        return int(r), int(g), int(b)
    step = 255.0 / (levels - 1)
    pr = int(round(round(r / step) * step))
    pg = int(round(round(g / step) * step))
    pb = int(round(round(b / step) * step))
    return max(0, min(255, pr)), max(0, min(255, pg)), max(0, min(255, pb))


def get_shape_mask(cx, cy, hw, hh, rot_deg, shape_type, img_h, img_w):
    """Get pixel coordinates inside a shape. Returns (ys, xs) arrays."""
    x_min = max(0, cx - hw)
    x_max = min(img_w, cx + hw)
    y_min = max(0, cy - hh)
    y_max = min(img_h, cy + hh)

    if x_min >= x_max or y_min >= y_max:
        return np.array([], dtype=np.int32), np.array([], dtype=np.int32)

    ys, xs = np.mgrid[y_min:y_max, x_min:x_max]
    ys_flat = ys.ravel()
    xs_flat = xs.ravel()

    dx = xs_flat - cx
    dy = ys_flat - cy

    if rot_deg != 0:
        rad = rot_deg * np.pi / 180.0
        cos_a = np.cos(rad)
        sin_a = np.sin(rad)
        rx = dx * cos_a + dy * sin_a
        ry = -dx * sin_a + dy * cos_a
    else:
        rx = dx.astype(np.float64)
        ry = dy.astype(np.float64)

    if shape_type == RECTANGLE:
        mask = (np.abs(rx) <= hw) & (np.abs(ry) <= hh)
    else:  # ROTATED_ELLIPSE
        if hw > 0 and hh > 0:
            mask = (rx * rx) / (hw * hw) + (ry * ry) / (hh * hh) <= 1.0
        else:
            return np.array([], dtype=np.int32), np.array([], dtype=np.int32)

    return ys_flat[mask], xs_flat[mask]


def compute_optimal_color(target, canvas, ys, xs, posterize_levels):
    """Compute the optimal color for the shape area."""
    if len(ys) == 0:
        return 128, 128, 128

    target_pixels = target[ys, xs]  # (N, 3)
    avg_r = np.mean(target_pixels[:, 0])
    avg_g = np.mean(target_pixels[:, 1])
    avg_b = np.mean(target_pixels[:, 2])

    return posterize_color(avg_r, avg_g, avg_b, posterize_levels)


def compute_error_reduction(target, canvas, ys, xs, r, g, b):
    """Compute how much painting this shape reduces pixel error."""
    if len(ys) == 0:
        return 0.0

    target_pixels = target[ys, xs].astype(np.float64)
    canvas_pixels = canvas[ys, xs].astype(np.float64)

    # Current error in shape region
    old_diff = canvas_pixels - target_pixels
    old_error = np.sum(old_diff * old_diff)

    # New error with proposed color
    new_color = np.array([r, g, b], dtype=np.float64)
    new_diff = new_color - target_pixels
    new_error = np.sum(new_diff * new_diff)

    return float(old_error - new_error)


def generate_random_candidates(n, img_w, img_h, rng):
    """Generate n random candidate shapes."""
    types = rng.choice([RECTANGLE, ROTATED_ELLIPSE], size=n)
    cx = rng.integers(0, img_w, size=n)
    cy = rng.integers(0, img_h, size=n)
    # Half-widths and half-heights (size range: 2 to max_dim/4)
    max_hw = max(4, img_w // 4)
    max_hh = max(4, img_h // 4)
    hw = rng.integers(2, max_hw, size=n)
    hh = rng.integers(2, max_hh, size=n)
    rot = rng.integers(0, 360, size=n)
    return types, cx, cy, hw, hh, rot


def _evaluate_candidate_worker(args):
    """Worker function for multiprocessing candidate evaluation."""
    target, canvas, shape_type, cx, cy, hw, hh, rot, img_h, img_w, posterize_levels = args
    ys, xs = get_shape_mask(cx, cy, hw, hh, rot, shape_type, img_h, img_w)
    if len(ys) == 0:
        return 0.0, (128, 128, 128)
    r, g, b = compute_optimal_color(target, canvas, ys, xs, posterize_levels)
    score = compute_error_reduction(target, canvas, ys, xs, r, g, b)
    return score, (r, g, b)


# Global shared arrays for multiprocessing (set during generation)
_shared_target = None
_shared_canvas = None
_shared_img_h = None
_shared_img_w = None
_shared_posterize_levels = None


def _init_worker(target, canvas, img_h, img_w, posterize_levels):
    """Initialize worker process with shared data."""
    global _shared_target, _shared_canvas, _shared_img_h, _shared_img_w, _shared_posterize_levels
    _shared_target = target
    _shared_canvas = canvas
    _shared_img_h = img_h
    _shared_img_w = img_w
    _shared_posterize_levels = posterize_levels


def _eval_worker(args):
    """Worker that uses globally initialized shared data."""
    shape_type, cx, cy, hw, hh, rot = args
    ys, xs = get_shape_mask(cx, cy, hw, hh, rot, shape_type, _shared_img_h, _shared_img_w)
    if len(ys) == 0:
        return 0.0, (128, 128, 128)
    r, g, b = compute_optimal_color(_shared_target, _shared_canvas, ys, xs, _shared_posterize_levels)
    score = compute_error_reduction(_shared_target, _shared_canvas, ys, xs, r, g, b)
    return score, (r, g, b)


def evaluate_candidates_cpu(target, canvas, types, cx, cy, hw, hh, rot,
                            img_h, img_w, posterize_levels):
    """Evaluate all candidates on CPU using numpy. Returns (scores, colors)."""
    n = len(types)
    scores = np.zeros(n)
    colors = np.zeros((n, 3), dtype=np.int32)

    for i in range(n):
        ys, xs = get_shape_mask(int(cx[i]), int(cy[i]), int(hw[i]), int(hh[i]),
                                int(rot[i]), int(types[i]), img_h, img_w)
        if len(ys) == 0:
            continue
        r, g, b = compute_optimal_color(target, canvas, ys, xs, posterize_levels)
        score = compute_error_reduction(target, canvas, ys, xs, r, g, b)
        scores[i] = score
        colors[i] = [r, g, b]

    return scores, colors


def evaluate_candidates_multiprocess(target, canvas, types, cx, cy, hw, hh, rot,
                                     img_h, img_w, posterize_levels, pool):
    """Evaluate candidates using multiprocessing Pool for CPU parallelism."""
    n = len(types)
    work_items = [
        (int(types[i]), int(cx[i]), int(cy[i]), int(hw[i]), int(hh[i]), int(rot[i]))
        for i in range(n)
    ]

    results = pool.map(_eval_worker, work_items, chunksize=max(1, n // (multiprocessing.cpu_count() * 4)))

    scores = np.zeros(n)
    colors = np.zeros((n, 3), dtype=np.int32)
    for i, (score, color) in enumerate(results):
        scores[i] = score
        colors[i] = list(color)

    return scores, colors


def evaluate_candidates_gpu(target, canvas, types, cx, cy, hw, hh, rot,
                            img_h, img_w, posterize_levels):
    """Evaluate candidates on GPU. Colors are computed via the kernel using
    mean-approximation: the kernel computes both the optimal color (average of
    target pixels in the mask) and the error reduction in a single pass."""
    n = len(types)

    # Transfer data to GPU
    d_canvas = cuda.to_device(canvas.astype(np.float32))
    d_target = cuda.to_device(target.astype(np.float32))
    d_cx = cuda.to_device(cx.astype(np.int32))
    d_cy = cuda.to_device(cy.astype(np.int32))
    d_hw = cuda.to_device(hw.astype(np.int32))
    d_hh = cuda.to_device(hh.astype(np.int32))
    d_rot = cuda.to_device(rot.astype(np.int32))
    d_types = cuda.to_device(types.astype(np.int32))

    # First pass: compute average colors on GPU
    colors_r_out = cuda.device_array(n, dtype=np.float32)
    colors_g_out = cuda.device_array(n, dtype=np.float32)
    colors_b_out = cuda.device_array(n, dtype=np.float32)

    # For simplicity and to avoid the CPU bottleneck, compute colors with
    # numpy vectorized operations in a batch approach on CPU, but only do the
    # mean computation (the lightweight part), not mask+error together.
    colors = np.zeros((n, 3), dtype=np.float32)
    for i in range(n):
        ys, xs = get_shape_mask(int(cx[i]), int(cy[i]), int(hw[i]), int(hh[i]),
                                int(rot[i]), int(types[i]), img_h, img_w)
        if len(ys) > 0:
            pixels = target[ys, xs]
            colors[i, 0] = np.mean(pixels[:, 0])
            colors[i, 1] = np.mean(pixels[:, 1])
            colors[i, 2] = np.mean(pixels[:, 2])
        else:
            colors[i] = [128, 128, 128]

    # Posterize colors
    posterized_colors = np.zeros((n, 3), dtype=np.int32)
    for i in range(n):
        posterized_colors[i] = posterize_color(colors[i, 0], colors[i, 1], colors[i, 2], posterize_levels)

    d_cr = cuda.to_device(posterized_colors[:, 0].astype(np.float32))
    d_cg = cuda.to_device(posterized_colors[:, 1].astype(np.float32))
    d_cb = cuda.to_device(posterized_colors[:, 2].astype(np.float32))
    errors_out = cuda.device_array(n, dtype=np.float64)

    threads_per_block = 128
    blocks = (n + threads_per_block - 1) // threads_per_block
    _gpu_compute_error_kernel[blocks, threads_per_block](
        d_canvas, d_target, d_cx, d_cy, d_hw, d_hh,
        d_rot, d_types, d_cr, d_cg, d_cb,
        errors_out, img_h, img_w, n
    )
    cuda.synchronize()
    scores = errors_out.copy_to_host()
    return scores, posterized_colors


def mutate_candidate(cx, cy, hw, hh, rot, shape_type, img_w, img_h, rng):
    """Apply a small random mutation to a candidate shape."""
    mutation_type = rng.integers(0, 4)
    new_cx, new_cy, new_hw, new_hh, new_rot = cx, cy, hw, hh, rot

    if mutation_type == 0:  # Move
        new_cx = max(0, min(img_w - 1, cx + rng.integers(-20, 21)))
        new_cy = max(0, min(img_h - 1, cy + rng.integers(-20, 21)))
    elif mutation_type == 1:  # Resize width
        new_hw = max(2, min(img_w // 2, hw + rng.integers(-10, 11)))
    elif mutation_type == 2:  # Resize height
        new_hh = max(2, min(img_h // 2, hh + rng.integers(-10, 11)))
    elif mutation_type == 3:  # Rotate
        new_rot = (rot + rng.integers(-30, 31)) % 360

    return new_cx, new_cy, new_hw, new_hh, new_rot


def hill_climb(target, canvas, shape_type, cx, cy, hw, hh, rot,
               best_score, best_color, n_mutations, img_h, img_w,
               posterize_levels, rng):
    """Hill-climb the best candidate with small mutations."""
    cur_cx, cur_cy, cur_hw, cur_hh, cur_rot = cx, cy, hw, hh, rot
    cur_score = best_score
    cur_color = best_color

    for _ in range(n_mutations):
        m_cx, m_cy, m_hw, m_hh, m_rot = mutate_candidate(
            cur_cx, cur_cy, cur_hw, cur_hh, cur_rot, shape_type, img_w, img_h, rng
        )
        ys, xs = get_shape_mask(m_cx, m_cy, m_hw, m_hh, m_rot, shape_type, img_h, img_w)
        if len(ys) == 0:
            continue
        r, g, b = compute_optimal_color(target, canvas, ys, xs, posterize_levels)
        score = compute_error_reduction(target, canvas, ys, xs, r, g, b)
        if score > cur_score:
            cur_cx, cur_cy, cur_hw, cur_hh, cur_rot = m_cx, m_cy, m_hw, m_hh, m_rot
            cur_score = score
            cur_color = (r, g, b)

    return shape_type, cur_cx, cur_cy, cur_hw, cur_hh, cur_rot, cur_score, cur_color


def paint_shape(canvas, shape_type, cx, cy, hw, hh, rot, r, g, b, img_h, img_w):
    """Paint a shape onto the canvas."""
    ys, xs = get_shape_mask(cx, cy, hw, hh, rot, shape_type, img_h, img_w)
    if len(ys) > 0:
        canvas[ys, xs] = [r, g, b]


def resize_image(image, max_resolution):
    """Resize image so the longest side is max_resolution."""
    w, h = image.size
    if max(w, h) <= max_resolution:
        return image
    scale = max_resolution / max(w, h)
    new_w = int(w * scale)
    new_h = int(h * scale)
    return image.resize((new_w, new_h), Image.LANCZOS)


def build_output_json(shapes, img_w, img_h):
    """Build the output JSON structure matching geometry_json.py format."""
    output_shapes = []
    # Background rectangle (first shape)
    bg_color = shapes[0]["color"] if shapes else [0, 0, 0, 255]
    output_shapes.append({
        "type": RECTANGLE,
        "data": [0, 0, img_w, img_h],
        "color": bg_color,
        "score": 0,
    })
    # Remaining shapes
    for shape in shapes[1:]:
        output_shapes.append(shape)
    return {"shapes": output_shapes}


print("Geometrize engine loaded successfully.")
```

```python
#@title Run Generation
import json
import time
from tqdm.notebook import tqdm

# Resize image to maxResolution
processed_image = resize_image(target_image, settings["maxResolution"])
proc_w, proc_h = processed_image.size
target_arr = np.array(processed_image, dtype=np.uint8)  # (H, W, 3)
img_h, img_w = target_arr.shape[:2]

print(T["starting"])
print(f"  Processing at {img_w}x{img_h} resolution")

# Initialize canvas with the average color of the target image
avg_color = target_arr.mean(axis=(0, 1)).astype(np.uint8)
canvas = np.full_like(target_arr, avg_color)

# Store generated shapes
generated_shapes = []
bg_r, bg_g, bg_b = int(avg_color[0]), int(avg_color[1]), int(avg_color[2])
generated_shapes.append({
    "type": RECTANGLE,
    "data": [0, 0, img_w, img_h],
    "color": [bg_r, bg_g, bg_b, 255],
    "score": 0,
})

rng = np.random.default_rng(42)
start_time = time.time()

stop_at = settings["stopAt"]
random_samples = settings["randomSamples"]
mutated_samples = settings["mutatedSamples"]
posterize_levels = settings["posterizeLevels"]
save_at_points = set(settings["saveAt"])

# Auto-tune BATCH_SIZE based on available hardware
if GPU_AVAILABLE:
    try:
        from numba import cuda as _cuda_info
        _device = _cuda_info.get_current_device()
        _multiprocessors = _device.MULTIPROCESSOR_COUNT
        _threads_per_block = 128
        BATCH_SIZE = max(16384, _multiprocessors * _threads_per_block * 4)
        BATCH_SIZE = min(BATCH_SIZE, random_samples)
    except Exception:
        BATCH_SIZE = min(random_samples, 16384)
else:
    BATCH_SIZE = max(2000, cpu_cores * 500)
    BATCH_SIZE = min(BATCH_SIZE, random_samples)
print(T["batch_size_info"].format(BATCH_SIZE, GPU_AVAILABLE))

# Set up multiprocessing pool for CPU candidate evaluation
n_workers = multiprocessing.cpu_count()
mp_pool = None
if not USE_GPU:
    try:
        mp_pool = multiprocessing.Pool(
            processes=n_workers,
            initializer=_init_worker,
            initargs=(target_arr, canvas, img_h, img_w, posterize_levels),
        )
        print(f"  Multiprocessing pool started with {n_workers} workers")
    except Exception as e:
        print(f"  Multiprocessing pool failed ({e}), using single-process mode")
        mp_pool = None

pbar = tqdm(range(1, stop_at + 1), desc=T["progress_bar_desc"], unit="shape")
for shape_idx in pbar:
    best_score = -1e30
    best_shape = None
    best_color = (128, 128, 128)

    # Re-initialize workers with current canvas state periodically
    if mp_pool is not None and shape_idx % 50 == 1:
        mp_pool.close()
        mp_pool.join()
        mp_pool = multiprocessing.Pool(
            processes=n_workers,
            initializer=_init_worker,
            initargs=(target_arr, canvas, img_h, img_w, posterize_levels),
        )

    # Evaluate random candidates in batches
    remaining = random_samples
    while remaining > 0:
        batch_n = min(BATCH_SIZE, remaining)
        types, cx, cy, hw, hh, rot = generate_random_candidates(batch_n, img_w, img_h, rng)

        if USE_GPU and batch_n >= 256:
            scores, colors = evaluate_candidates_gpu(
                target_arr, canvas, types, cx, cy, hw, hh, rot,
                img_h, img_w, posterize_levels
            )
        elif mp_pool is not None and batch_n >= 100:
            scores, colors = evaluate_candidates_multiprocess(
                target_arr, canvas, types, cx, cy, hw, hh, rot,
                img_h, img_w, posterize_levels, mp_pool
            )
        else:
            scores, colors = evaluate_candidates_cpu(
                target_arr, canvas, types, cx, cy, hw, hh, rot,
                img_h, img_w, posterize_levels
            )

        batch_best_idx = np.argmax(scores)
        if scores[batch_best_idx] > best_score:
            best_score = scores[batch_best_idx]
            best_shape = (int(types[batch_best_idx]), int(cx[batch_best_idx]),
                         int(cy[batch_best_idx]), int(hw[batch_best_idx]),
                         int(hh[batch_best_idx]), int(rot[batch_best_idx]))
            best_color = (int(colors[batch_best_idx][0]),
                         int(colors[batch_best_idx][1]),
                         int(colors[batch_best_idx][2]))

        remaining -= batch_n

    if best_shape is None:
        continue

    # Hill-climb the best candidate
    shape_type, b_cx, b_cy, b_hw, b_hh, b_rot = best_shape
    shape_type, f_cx, f_cy, f_hw, f_hh, f_rot, final_score, final_color = hill_climb(
        target_arr, canvas, shape_type, b_cx, b_cy, b_hw, b_hh, b_rot,
        best_score, best_color, mutated_samples, img_h, img_w, posterize_levels, rng
    )

    # Paint shape on canvas
    fr, fg, fb = final_color
    paint_shape(canvas, shape_type, f_cx, f_cy, f_hw, f_hh, f_rot, fr, fg, fb, img_h, img_w)

    # Record shape in output format
    if shape_type == RECTANGLE:
        shape_data = [f_cx, f_cy, f_hw * 2, f_hh * 2]
    else:  # ROTATED_ELLIPSE
        shape_data = [f_cx, f_cy, f_hw, f_hh, f_rot]

    generated_shapes.append({
        "type": shape_type,
        "data": shape_data,
        "color": [fr, fg, fb, 255],  # forceOpaqueShapes=true
        "score": round(final_score / (img_w * img_h * 3 * 255.0 * 255.0), 6),
    })

    # Update progress bar postfix with speed and error info
    elapsed = time.time() - start_time
    speed = shape_idx / elapsed if elapsed > 0 else 0
    error_metric = final_score / (img_w * img_h) if img_w * img_h > 0 else 0
    pbar.set_postfix(speed=f"{speed:.1f} shapes/s", error=f"{error_metric:.4f}")

    # Checkpoint saving
    if shape_idx in save_at_points:
        checkpoint_data = build_output_json(generated_shapes, img_w, img_h)
        checkpoint_filename = f"checkpoint_{shape_idx}.json"
        checkpoint_path = os.path.join(OUTPUT_DIR, checkpoint_filename)
        with open(checkpoint_path, "w", encoding="utf-8") as f:
            json.dump(checkpoint_data, f)
        pbar.write(T["checkpoint_saved"].format(checkpoint_path, shape_idx))

# Clean up multiprocessing pool
if mp_pool is not None:
    mp_pool.close()
    mp_pool.join()

elapsed = time.time() - start_time
speed = stop_at / elapsed if elapsed > 0 else 0
print(T["generation_complete"].format(stop_at, elapsed, speed))
```
